## First we need all necessary imports

```python
import os
import numpy as np
import flopy.modflow as mf
import shutil
```

* The modflow-files will be stored in './workspace' 
* Each calculation will delete the files

```python
workspace = os.path.join('rp1')

if os.path.exists(workspace):
    shutil.rmtree(workspace)
    
if not os.path.exists(workspace):
    os.makedirs(workspace)
```

### Setup Model

```python
model = mf.Modflow(modelname='RP1', exe_name='mf2005', model_ws=workspace)
```

### Setup Discretization and Basic Package

```python
nrow = 12
ncol = 11
nlay = 3

delr = 11000/ncol
delc = 6000/nrow

nper = 13
perlen = [1, 30, 28, 31, 30, 31, 30 , 31, 31, 30, 31, 30, 31]
steady = [True, False, False, False, False, False, False, False, False, False, False, False, False]

dis = mf.ModflowDis(model, nlay=nlay, nrow=nrow, ncol=ncol, perlen=perlen, delr=delr, delc=delc, top=200, botm=[180, 170, 80], nper=nper, steady=steady)
dis.plot()
```

```python
bas = mf.ModflowBas(model, ibound=1, strt=200)
bas.plot()
```

### Setup Wells

```python
spd = {
    0: [
        [2, 4, 4, 0],
    ],
    1: [
        [2, 4, 4, -15000],
    ]
}
wel = mf.ModflowWel(model, stress_period_data=spd)
wel.plot()
```

### Setup Flow-Package (incl. soil parameters)

```python
laytyp=[1, 0, 0]

hk = [100, 0.01, 200]
chani = 0.5

chani=1.0
layvka=[1, 1, 1]
vka=[0.1, 1, 0.1]

ss=[1e-05, 1e-5, 2e-5]
sy=[0.15, 0.15, 0.1]

lpf = mf.ModflowLpf(model, laytyp=laytyp, hk=hk, chani=chani, layvka=layvka, vka=vka, ss=ss, sy=sy)
lpf.plot()
```

### The Solver and Output Control

```python
pcg = mf.ModflowPcg(model, rclose=1e-2, hclose=1e-2)

spd = {}
for i in range(nper):
    spd[(i, 0)] = ['save head', 'save budget']
        
oc = mf.ModflowOc(model, stress_period_data=spd)
```

### Calculate the model

```python
model.write_input()
model.run_model()
```
